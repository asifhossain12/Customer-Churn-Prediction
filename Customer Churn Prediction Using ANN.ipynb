```python
# Label Encoding 
def print_unique_col(df1):
    for column in df1:
        if df1[column].dtypes=='object':
            print(f'{column} : {df1[column].unique()}')
```

```python
print_unique_col(df1)
```

```python
df1.tail()
```

```python
df1.replace('No internet service', 'No', inplace=True)
df1.replace('No phone service', 'No', inplace=True)
```

```python
print_unique_col(df1)
```

```python
# Replace the Yes No columns with 0's and 1s 

yes_no_col = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for col in yes_no_col:
    df1[col].replace({'Yes': 1, 'No': 0}, inplace=True)
```

```python
print_unique_col(df1)
```

```python
for column in df1:
        print(f'{column} : {df1[column].unique()}')
```

```python
df1['gender'].replace({'Female': 1, 'Male': 0}, inplace=True)
```

```python
df1['gender'].unique()
```

```python
df2 = pd.get_dummies(data=df1, columns=['InternetService', 'Contract', 'PaymentMethod']) 
df2[df2.columns] = df2[df2.columns].astype(int)
df2.columns
```

```python
df2.sample(4)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
#Standardizing 
cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges'] 

from sklearn.preprocessing import MinMaxScaler 

scaler = MinMaxScaler() 

df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale]) 
```

```python
df2.sample(5)
```

```python
for col in df2:
    print(f'{col} : {df2[col].unique()}')
```

```python
X = df2.drop('Churn', axis='columns')
y = df2['Churn']
```

```python
# Train and Test Data 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=5, test_size=0.2)
```

```python
X_train.shape
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf 
from tensorflow import keras 
```

```python
model = keras.Sequential([
    keras.layers.Dense(20, input_shape=(26,), activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'), 
]) 

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
              
model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python

y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```
